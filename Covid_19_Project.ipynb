```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

**1.Read csv**

```python
df=pd.read_csv('Post Covid-19.csv',encoding='iso-8859-1', quotechar='"')
df.info()
df
```

**2.Remove Duplicates**

```python
df.drop_duplicates(inplace=True)
df
```

**3.Changing Data Type**

```python
df['Time Period Start Date']=pd.to_datetime(df['Time Period Start Date'],errors='coerce')

df['Time Period End Date']=pd.to_datetime(df['Time Period End Date'],errors='coerce')

df.info()
```

**4.Delete Unwanted Columns**

```python
df.drop(['Time Period Label','Suppression Flag','Confidence Interval','Quartile range'],axis=1,inplace=True)
df
```

**5.Handling Missing Values**

```python
df.isnull().sum()
df['Value'].fillna(df['Value'].mean(), inplace=True)
df['LowCI'].fillna(df['LowCI'].mean(), inplace=True)
df['HighCI'].fillna(df['HighCI'].mean(), inplace=True)
df['Quartile number'].fillna(df['Quartile number'].mean(), inplace=True)

df.info()
```

**6.Visualize**

```python
sns.histplot(df['Time Period'],kde=True)
plt.title('Time Period of Patients')
plt.show()
```

**7.By Group**

```python
sns.lineplot(data=df[df['Group']=='National Estimate'], x='Time Period Start Date', y='Value')
plt.title('National Long COVID Trend Over Time')
```

**8.By age and subgroup**

```python
sns.barplot(data=df[df['Group']=='By Age'], x='Subgroup', y='Value', palette='coolwarm')
plt.xticks(rotation=45)
plt.title('Long COVID Percentage by Age Group')
```

**9.By state**

```python
sns.barplot(data=df[df['Group']=='By State'], x='State', y='Value')
plt.xticks(rotation=90)
plt.title('Long COVID Rate by State')
```

**10.By Gender**

```python
sns.barplot(data=df[df['Group']=='By Sex'], x='Subgroup', y='Value', palette='pastel')
plt.title('Long COVID by Gender')
```

**11.Heatmap**

```python
pivot = df.pivot_table(values='Value', index='State', columns='Time Period', aggfunc='mean')
sns.heatmap(pivot, cmap='YlOrRd')
plt.title('Heatmap of Long COVID Rates by State and Time')
```

**12.Histplot**

```python
sns.histplot(df['Value'], bins=20, kde=True, color='teal')
plt.title('Distribution of Long COVID Percentages')
```

**13.Quartile Analysis**

```python
sns.boxplot(data=df, x='Quartile number', y='Value', palette='viridis')
plt.title('Quartile-wise Long COVID Rates')
```
