```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

**1. Read csv**

```python
df=pd.read_csv('Sales.csv',encoding='iso-8859-1')
df.info()
df
```

**2.Rename Columns**

```python
new={'Postal Code':'Pin Code','Sub-Category':'Sub Category'}
df.rename(columns=new,inplace=True)
df
```

**3.Update Column Values**

```python
new={'Furniture':'Furniture & Sofa','Technology':'Technology & Gadgets'}
df['Category']=df['Category'].map(new)
df
```

**4 Remove Unwanted Characters**

```python
df['Customer Name']=df['Customer Name'].str.strip()
df
df['Product Name']=df['Product Name'].str.strip()
df
```

**5.Change Data Type**

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df.info()
```

**6.Replace Specific Values**

```python
df['Category']=df['Category'].fillna(method='ffill')
df
```

**7. Drop Duplicates**

```python
df=df.drop_duplicates()
df
```

**8. Apply String Function**

```python
df['Customer Name']=df['Customer Name'].str.capitalize()
df
```

9.Feature Engineering

```python
df['Delivery Days'] = (df['Ship Date']-df['Order Date']).dt.days

df['Profit Margin'] = df['Profit'] / df['Sales'] * 100

df
```

**10.Visualize the count of males and females based on gender column using seaborn**

```python
a=sns.countplot(x='Category',data=df)
a.set_facecolor('lightgreen')
for i in a.containers:
  a.bar_label(i)
```

**11.Group Based on Category**

```python
ge=df.groupby('Category')['Sales'].sum()
d=pd.DataFrame(ge)

b=sns.barplot(x='Category',y='Sales',data=d)
b.set_facecolor('lightgreen')
for i in b.containers:
  b.bar_label(i,fmt='%0.0f')
```

**12.Based on Quantity And Sales**

```python
ge=df.groupby(['Quantity','Category'])['Sales'].sum()
d=pd.DataFrame(ge)
plt.figure(figsize=(10,5))

c=sns.barplot(x='Quantity',y='Sales',data=d,hue='Category')
c.set_facecolor('lightgreen')
for i in c.containers:
  c.bar_label(i,fmt='%d')
```

13. Based on Sub Category

```python
ge=df.groupby(['Sub Category','Category'])['Sales'].sum()
d=pd.DataFrame(ge)
plt.figure(figsize=(10,20))

d=sns.barplot(x='Sub Category',y='Sales',data=d,hue='Category')
d.set_facecolor('lightgreen')
for i in d.containers:
  d.bar_label(i,fmt='%d')
```
